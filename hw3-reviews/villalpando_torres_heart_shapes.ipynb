## **“Heart representations with heart shapes” : Which conditions or pathologies are the most correlated with heart shapes?**


 **I. Introduction**


Heart disease is the leading cause of death in the United States. As of 2021, out of 659,000 deaths about 1 in 4 can be attributed to a heart condition (CDC). Health care services, medicines, and lost productivity attributed to heart disease costs the United States over 363 billion dollars. 

Early detection is key to preventing deaths attributed to cardiovascular disease. If detected early enough, there are pharmacological and lifestyle interventions that can be made to slow the progression of the disease and improve the patient’s quality of life (Wang et al, 2015). The bottleneck in detecting heart disease during early stages has been that there are currently no robust means of detecting heart disease. Advances in the way medical data is stored and shared combined with the adoption of machine learning strategies have paved a path to predict and classify heart disease.

The heart is composed of walls, chambers,valves, vessels, and an electrical conduction system. The work presented here focuses on observing changes in the ventricles (both left and right) and changes in the structure of the middle, muscular, myocardial wall. The right ventricle is responsible for pumping deoxygenated blood into the lungs and the left ventricle pumps blood to the rest of your body. 

The heart is an incredibly dynamic organ that remodels throughout development and in the presence of disease. Specifically in disease, changes such as increasing/decreasing ventricle size and/or increasing/decreasing thickness of the myocardial wall. With the assumption that these changes are significant enough to cause changes in the overall geometry of the heart, Machine Learning and statistics have the power to make predictions on heart disease. Here, I review 5 different papers that utilize the power of Machine Learning and statistics to predict and classify heart disease. Together, the work presented here is making small steps in achieving translational impact in clinical medicine. This work is quickly evolving, and will likely have a huge impact on the way clinicians diagnose their patients in the near future.


**II. Literature review**

**An ensemble of neural networks provides expert-level prenatal detection of complex congenital heart disease (Arnaout et al, 2021)**

Motivated by the high survival rate of fetuses whose Congenital heart disease (CHD) was detected at an early enough stage to enable in utero surgery, Arnout et al sought to use a mixture of Deep Learning(DL) techniques to improve ultrasound analysis for CHD. As part of routine prenatal care, expectant mothers are encouraged to take an ultrasound during the second trimester of pregnancy. This ultrasound routinely involves imaging 5 different sections of the heart (Figure 1A). These ultrasound scans were used to distinguish between healthy hearts and hearts with CHD. 
Methods
Five test datasets from UCSF and Boston Children’s Hospital were used to test the model. The model consisted of a binary ‘DL dx’ classifier to assign the hearts to a healthy or a CHD class as well as deep learning segmentation to calculate the cardiothoracic measurements of the fetal heart. 

DL dx Classifier: A convolutional neural network was trained to pick out the 5 screening views (3VT, 3VV, LVOT, A4O,ABDO) from fetal ultrasounds. That same network was then trained to classify normal hearts vs CHD for each view based on a rule-based classifier that was comparing weighted averages between the two classes (normal vs. CHD). The clinician’s diagnosis of the scan was used as the gold-standard to verify model predictions. On average, the model’s specificity(96%) and sensitivity(95%) were similar to that of the clinician’s. They further validated the model using data that was lower quality and data collected from a multifetal pregnancy (twins).  
 
Segmentation for fetal biometrics: A U-net was trained such that it would be able to segment apart the different cardiothoracic segments of the heart using A4C. From this, the model was able to measure the cardiothoracic ratio, cardiac axis, and fractional area change of each chamber of the heart. These biometric values were compared in CHD vs healthy and there appeared to be a clear difference in cardiac axis size between the two groups. 

This model opens a path for clinicians to be able to detect CHD at an early stage and in doing so allows for early clinical intervention. Further training with datasets of varying quality from around the world would make this model more robust. This work fits into the topic of heart representations with heart shapes because it is classifying a heart as “healthy” or “diseased” based solely on the different shapes and dimensions that it has in either condition. Shape segmentation strategies were implemented to measure variables such as cardiothoracic ratio, cardiac axis, and fractional area change for each chamber. Further, this work answers the question “which conditions or shapes are the most correlated with heart shapes” in that a disease such as CHD alters the geometry of the heart and can thus be diagnosed by comparing dimensions of a fetal heart to that of a healthy one. This relates to the other papers presented here in that it uses machine learning strategies, in this case specifically deep learning, to diagnose a heart as diseased or healthy.

**Cardiac magnetic resonance image-based classification of the risk of arrhythmias in post- myocardial infarction patients (Kotu et al, 2015)** 

Surviving a heart attack puts some survivors at a high risk of developing arrhythmias that require the insertion of a cardioverter defibrillator. For survivors with a lower risk of developing arrhythmias, the insertion results in unnecessary complications. Thus, there is a need to effectively determine whether a patient is a high or low risk patient. Here, Kotu et al use MRI images to determine left ventricular ejection volume as well as the myocardial scar size to predict if a patient is at high or low risk of developing an arrhythmia. A larger scar is related to a higher risk of developing an arrhythmia, and its location as well as its texture have been proposed to have some implications as well. Methods to assess this risk involved the use of classification schemes from MATLAB, k-nearest neighbor (k-NN), support vector machine (SVM), decision tree, and random forest. 

Methods

The dataset was obtained from the Department of Cardiology at Stavanger University Hospital. Here they only used images of the short axis of the myocardium where scars were visible. Segmented images of the myocardium (Figure 2) were then used to gather information such as scar volume, localization features, scar gradient, local binary patterns, and chi-square features. To determine the localization features of the myocardial scar a reference point CT was chosen with which the localization features could be compared to at a given slice (Figure 3). Phi was the angle made between the left ventricle and any arbitrary point on the myocardium.

Myocardial tissue was segmented into healthy and scarred by cardiologists, and 17 features were extracted from the scarred myocardial tissue. To determine which features were better at determining the  likelihood of developing an arrhythmia, a  leave-one-out cross-validation scheme from MATLAB was used to classify high and low risk groups. This first experiment resulted in a 94.44% accuracy and considered size, location, and heterogeneity of scar. In a second experiment, k-NN, SVM, a decision tree and a random forest were used to do nested cross-validation. This second experiment resulted in 92.6% average accuracy and took into account scar location and heterogeneity.  Overall, this model showed that scar texture was a critical factor in determining high or low risk of developing an arrhythmia and paves the way to enable cardiologists to make more reliable predictions.

This work fits into the topic of heart representations with heart shapes because it is classifying a heart as “high arrhythmic risk” or “low arrhythmic risk” based solely on the different shape and texture that the myocardium has in either condition. Further, this work answers the question “which conditions or shapes are the most correlated with heart shapes” in that the severity of a heart attack can change the texture of the heart, and this change in texture can be used to predict the likelihood that a patient will develop an arrhythmia. This relates to the other papers presented here in that it uses machine learning tools to predict patient outcomes and with hopes of improving the patient’s quality of life.

**Fast and accurate view classification of echocardiograms using deep learning (Madani et al, 2018)**

Echocardiography is one of the most widely used methods to diagnose and evaluate cardiovascular function. Echocardiography is made up of videos and still images, which can make it difficult to decipher what is in view. In order to make accurate diagnoses using echocardiography, there needs to be some knowledge about what structure(s) are in view. Motivated by recent innovations in applying deep learning techniques to enable image-based diagnosis, Madani et al sought to apply these techniques to the classification of echocardiogram views. They acknowledge that a bottleneck in the application of machine learning techniques to enable view classification has been hindered by the time consuming manual selection and annotation of figures. Here they use supervised deep learning with convolutional neural networks (Figure 4)  to classify views and they are able to achieve 91.7% accuracy. 

Methods

Randomly selected electrocardiograms were used to train the model, and an entirely different dataset was used to test its performance. These electrocardiograms was made up of patients with pathologies related to changes in variables such as left ventricular mass index, left ventricular ejection volume, Left Ventricular End-Systolic Volume, right ventricular systolic pressure, and Tricuspid annular plane systolic excursion. Interestingly, this group also made sure that the data collected consisted of different qualities as a result of differences in data acquisition. To further ensure translational impact, a small training set was used to achieve high accuracy. To address the ambiguity that deep learning entails, and to better understand which features the program is using to classify each view, the authors performed occlusion experiments to assess the models performance when clinically relevant features (i.e. chambers or valves) were obstructed. Interestingly, they found that the model uses the same features that a trained eye would use to make their decision. 

This work fits into the topic of heart representations with heart shapes because it is able to classify an image of a heart to a certain “view” that could be imaged based solely on the different shapes that the model is seeing. Further, this work answers the question “which conditions or shapes are the most correlated with heart shapes” in that it appears that different diseases alter the geometry of the heart in unique ways that allow an algorithm to predict what part of the heart it is looking at. This relates to the other papers presented here in that it uses machine learning strategies, in this case specifically deep learning, to classify an image to belong to a certain “view” of the heart.

**Deep echocardiography: data-efficient supervised and semi-supervised deep learning towards automated diagnosis of cardiac disease (Madani et al, 2018)**

After establishing a convolutional neural network architecture to be able to classify various heart views, Madani et al adapted their model such that it would be able to achieve binary left ventricular hypertrophy(LVH) classification (Figure 5).  This was made possible because certain views of the heart make it easier to observe if there is enlargement of chambers or ventricular walls. The big idea here is that if an algorithm can learn what anatomical structure it is looking at based solely on characteristics pertaining to its shape, then it should be able to learn patterns and classify things as normal or not normal. Using a large training dataset they were able to train the model with an ensemble of CNNs and a U-net to achieve view segmentation with a 94.4% accuracy (supervised learning). When training the same model using only a limited training dataset they were only able to achieve 91.2% accuracy. To address this issue a generative adversarial network (GAN) was used and they were able to achieve a 92.3% accuracy in classifying left ventricular hypertrophy. 

Methods

The dataset used was obtained from UCSF. The dataset consisted of echocardiography scans of varying quality. Images were downsized to ensure that they each had a 120 x 160 resolution. To remove extraneous data that was less relevant to the classification, relevant structures within the apical 4c chamber were segmented with a U-net approach. The field in view was classified using a convolutional neural network, and from there images that were classified as being a4c view were LVH positive or negative. 

This model is exciting because it is able to take echocardiograms of varying qualities and enables clinicians to make a patient's diagnosis faster. It is able to do this because it can detect the field of view quickly, a task that takes a human much longer. This work fits into the topic of heart representations with heart shapes because it is able to classify an image of a heart to a certain “view” that could be imaged based solely on the different shapes that the model is seeing. Further, this work answers the question “which conditions or shapes are the most correlated with heart shapes” in that it appears that a pathology such as left ventricular hypertrophy is detectable just by looking at changes in geometry of the heart. This relates to the other papers presented here in that it uses machine learning strategies, in this case specifically deep learning, to classify an image as having a disease, and this group builds off of the model that they had developed only months earlier.

**A Statistical Model for Quantification and Prediction of Cardiac Remodeling: Application to Tetralogy of Fallot (Mansi et al, 2011)**

Repaired Tetralogy of Fallot (rTOF) is a heart disease that affects the right ventricle and pulmonary arteries that requires surgical repair. However, intervention this late into the disease progression can result in damage to the pulmonary valves and in turn lead to complications such as altered electrophysiology and severe right ventricle dilation. If a pulmonary valve replacement(PVR) is needed, it can reduce the risk of complications and improve the quality of life for the patient, however it is important to get the timing of the PVR perfect. PVR that is too early could lead to implant dysfunction and PVR that is too late has no effect since the right ventricle could potentially be too damaged at that point. Thus, it is important to have a way to analyze and quantify changes in the structure of the right ventricle to be able to effectively treat the condition. The main goal of this paper was to try to estimate how the right ventricle would remodel over time using computational shape analysis. 

Methods

Here, cine MRI images obtained from three different hospitals were used to train and test the model. The right ventricle is segmented from the model and shape analysis was performed to create a template for a right ventricle. A forward approach was used such that the template takes into account the average of all the deformed observations as well as residuals. Different landmarks were then selected to identify pathological shape patterns. PCA analysis was done to determine which features were more likely to be changed in a disease state. Partial least squares (PLS) regression, which combines PCA and linear regression, and canonical correlation analysis were used to create a 3D model of the growth of the right ventricle (Figure 6).  

This work is particularly exciting because it could be applied to many other diseases and potentially save lives. This work fits into the topic of heart representations with heart shapes because it is able to predict what a patient’s right ventricle will look like after a certain amount of time. This enables clinicians to make more informed decisions about whether or not to perform PVR on a patient. Further, this work answers the question “which conditions or shapes are the most correlated with heart shapes” in that the severity of a disease such as Tetralogy of Fallot is very much associated with what the shape of the right ventricle is. This relates to the other papers presented here in that it uses different geometries of the heart to make predictions about the disease. It is different however, in that it isn’t necessarily relying on machine learning strategies, and is instead relying on statistics to make a prediction on how the shape will change over time.

**III. Conclusion.**

Here, I reviewed 5 different papers that looked at how changes in the shape of the heart could be leveraged to make predictions about its overall functional state. Most of these models relied on echocardiography data, and one relied on MRI data. Although each paper had very different goals, the one theme they all had in common was that they leveraged the power of machine learning tools (and statistics) to make predictions based on geometrical features of the heart. None of the models presented here were perfect - some were trained using “textbook quality” data and others had a lower accuracy score than others. However, work continues to build off of current advancements and will likely continue to address these issues. Fine tuning of these models will enable them to become accessible to clinicians worldwide and address challenges in image-based diagnostics.
