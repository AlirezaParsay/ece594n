### Introduction

Dimensinality reduction is well understood on euclidian distances when we want to estimate linear components from a higher dimensional space. 

However, there are several computer vision problems where it is more natural to consider features that
live in a non-Euclidean space like Grassmann manifolds and Lie groups are used for motion segmentation and multibody factorization.

This brings a need to develop a framework for dimensinality reduction and clustering data lying in different submanifolds of a Riemannian space.

### Background

One can extend nonlinear dimensionality reduction (NLDR) methods (often designed for one submanifold) to deal with multiple submanifolds. For instance, "R. Souvenir and R. Pless. Manifold clustering. In ICCV, 2005" combines Isomap "J. B. Tenenbaum et al. A global geometric framework for nonlinear dimensionality reduction. Science, 290(5500):2319â€“2323, 2000 "with EM.  

However , all these manifold clustering algorithms assume that the feature vectors are embedded in a Euclidean space and use the Euclidean metric or a variation of it to perform dimensinality reduction and clustering.

In this notebook we calculate the Local linear embeddding (LLE) on a manifold, which is one of the non linear dimensionality reduction (NLDR) techniques. 

We evaluate the LLE for 2 examples of data, one on a hypersphere of 2-dimensions and another on the sklearn's toy dataset.

For the first example, we generate 2 clusters of data and classify them in thier reduced dimensionality.

LLE consists of 3 main steps:
1. Find k-nearest neighbors
    - find the kNN of each datapoint $x_{i} $ by incorporating the Riemannian distance
2. Find the optimal weight matrix $W$ which minimizes the following reconstruction error:
    $$
    \varepsilon_{\operatorname{Riem}}(W)=\sum_{i=1}^{n}\left\|\sum_{j=1}^{n} W_{i j} \log _{\boldsymbol{x}_{i}}\left(\boldsymbol{x}_{j}\right)\right\|_{\boldsymbol{x}_{i}}^{2}
    $$
3. Estimate low dimensional vectors using the Sparse eigenvalue problem:
    - perform an eigenanalysis of $\widetilde{\textbf{M}}$ and obtain the $ (d + 1) $ eigenvectors associated with the smallest $(d + 1)$ eigenvalues

We can  then cluster the data into m groups by applying classification algorithms to the columns of a matrix whose rows are the $m$ eigenvectors in the null space of  $\widetilde{\textbf{M}}$

### Example 1:

Let us generate data on a hypersphere of 2 dimensions. To generate the data, we will take a point on the hypersphere  and project it along a tangent vector. The points along the geodesic, are considered as our input data.

Parameters:
- Manifold : S-curve
- Metric : Hypersphere 
- Samples : 200
- Input dimension : $\textbf{SO}^{3}$
- Output dimension : $\textbf{R}^{m} = \textbf{R}^{2}$
- Number of nearest neighbors : 12
- Number of clusters : 2 

#### Generate data  on the manifold:

```python
# Construct a sequence of points on the hypersphere
sphere = Hypersphere(dim=2)
metric = sphere.metric
n_samples = 200
```

```python
# Generate cluster 1

point = np.array([-0.65726771, -0.02678122, 0.7531812])
tangent_vector = sphere.to_tangent(np.array([2, 0, 0.8]), base_point=point)
result = sphere.metric.geodesic(initial_tangent_vec=tangent_vector, initial_point=point)
points_on_geodesic = result(np.linspace(0.0, 1.0, int(n_samples/2)))
```

```python
# Generate cluster 2

point2 = np.array([0.7557272951, 0.1955458503, -0.6250104606])
tangent_vector = sphere.to_tangent(np.array([1, -1, -1]), base_point=point2)
result = sphere.metric.geodesic(initial_tangent_vec=tangent_vector, initial_point=point2)
points_on_geodesic2 = result(np.linspace(0.0, 1.0, int(n_samples/2)))
```

Let's plot the generated points on the hypershpere.

The colors represent data in different clusters

```python
fig = plt.figure(figsize=(10, 10)); ax = fig.add_subplot(111, projection="3d")

ax = viz.plot(points_on_geodesic, ax=ax, space="S2", color="gold", label="Class A")
ax = viz.plot(points_on_geodesic2, ax=ax, space="S2", color="purple", label="Class B")
ax.legend();
```

Now, we perform LLE on the above generated sythetic data. 

LLE consists of 3 main steps:
1. Find k-nearest neighbors
2. Find the optimal weight matrix
3. Estimate low dimensional vectors

Let's calculate the k-nearest neighbours according to the manifold distance:

```python
# %%
# DIMENSIONALITY REDUCTION
# Local linear embedding
data = np.append(points_on_geodesic, points_on_geodesic2 ,axis= 0)
n_neighbors = 12  # neighborhood which is used to recover the locally linear structure
n_dimensions = 2 # number of coordinates for the manifold
n_classes = 2 # Number of clusters
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans( n_classes)
kmeans.fit(data)
labels = kmeans.predict(data)

k_nearest_vals = Compute_neighbours(data, labels, metric, n_neighbors)
```

After finding the k-nearest neighbors, we define the data points $x_{i}$ in terms of the local linear relationships within the neighborhood. For each $x_{i}$ in dimension $d$, we reconstruct $x_{i}$  using a linear combination of its k-nearest neighbors, using a weight matrix $ W $

We compute the weights matrix $ W $, which minimizes the Riemannian reconstruction error :
$$
\varepsilon_{\operatorname{Riem}}(W)=\sum_{i=1}^{n}\left\|\sum_{j=1}^{n} W_{i j} \log _{\boldsymbol{x}_{i}}\left(\boldsymbol{x}_{j}\right)\right\|_{\boldsymbol{x}_{i}}^{2}
$$

```python
M, W = Compute_W(data, metric ,k_nearest_vals, n_samples, n_neighbors )
```

To validate the weights obtained from the above optimization problem, we calculate the linear reconstruction $ \hat{\bar{\text{x}}} = \bar{\text{x}} W$ 

```python
Validate_W(data, W, k_nearest_vals)
```

From the above figure, we observe that the linearly reconstructed points from the k-nearest weighted neighbours are very close to the original data points $\bar{\text{x}} $ represented in various dimensions

 Hence using the weight matrix $W$ , we obtain a local linear reconstruction of the original data

Now, we solve for the lower dimensional embedding coordinates $y_{i} \in \textbf{R}^{m} $ that preserve local linear relationships.

 We perform an eigenanalysis of $\widetilde{\textbf{M}}$ and obtain the $(d + 1)$ eigenvectors associated with the smallest $(d + 1) $eigenvalues.
 The $d$ eigenvectors of the matrix $\widetilde{\textbf{M}}$ associated with its second to $(d + 1)$-th smallest eigenvalues correspond to eigenvectors spanning a d-dimensional space which contains the embedding coordinates

```python
optim = Optimize_y(M, n_samples,n_dimensions)
y_vector = np.reshape(optim.mimimize().x,(n_samples, n_dimensions))
```

```python
print('Reconstructed values : Y_hat')
print(y_vector[:3,:])
```

```python
embedding, _  = null_space(M,n_dimensions)
result = embedding
```

Now we split the reduced dimensionality data into test and train datasets. 

We will use the train data to train the k-NN classifier and test the performance using the test set.

```python
from sklearn.model_selection import train_test_split

knn = KNearestNeighborsClassifier(n_neighbors= n_neighbors, distance=metric.dist)
X_train, X_test, y_train, y_test = train_test_split(result,labels,train_size=0.8)

knn.fit(X_train, y_train)
y_pred =knn.predict(X_test)
print('Confusion matrix:')
emp_probability(n_classes, y_pred , y_test)

plot_2d(result, labels, 'LLE')
```

We can see that the shape of the data on the submanifold has clearly been split into different clusters which are easy to classify.

By applying nearest neigbours classification algorithm on the test data, we can see that all the test-data points are correctly classified according to thier labels.

### Example 2:

Now, lets try the local linear embedding technique on a toy dataset from the `sklearn.datasets` package.

We generate 200 points on the `S-curve` manifold in $\textbf{R}^{3}$

Let us apply LLE to this curve to reduce it to $\textbf{R}^{2}$

Parameters:
- Manifold : S-curve
- Metric : Euclidian 
- Samples : 500
- Input dimension : $\textbf{R}^{3}$
- Output dimension : $\textbf{R}^{2}$
- Number of nearest neighbors : 15

```python
# Test on toy dataset

from sklearn import manifold, datasets

rng = np.random.RandomState(0)

n_samples = 500
S_points, S_color = datasets.make_s_curve(n_samples, random_state=rng)
plot_3d(S_points, S_color, "Original S-curve samples")
```

The data points on the S-curve manifold, lie as shown above. 

Now we calculate the 12 nearest neighbors of the data points 

```python
n_neighbors = 15  # neighborhood which is used to recover the locally linear structure
n_components = 2  # number of coordinates for the manifold
```

```python
from geomstats.geometry.euclidean import Euclidean

coords = Euclidean(dim=3)
metric = coords.metric

from sklearn.cluster import KMeans

data = S_points
kmeans = KMeans( 1)
kmeans.fit(data)
labels = kmeans.predict(data)

k_nearest_vals = Compute_neighbours(data, labels, metric, n_neighbors)
```

```python
# Compute W
from utils import *
M, W = Compute_W(data, metric ,k_nearest_vals, n_samples, n_neighbors )
```

```python
embedding, _  = null_space(M,2)
result = embedding
```

```python
plot_2d(result, S_color, 'LLE')
```

From the above graph, we see the Local linear embedding of the toy dataset in 2 dimensions. We can observe that the gradient of colors are still preserved, implying that low dimensional representation preserves local linearity in LLE.

### Conclusion

We implemented the non-linear dimensionality reduction technique : Local linear embedding which assumes that the local neighborhood of a point on the manifold can be approximated by the affine subspace spanned by the k-nearest neighbors of the point and finds a low-dimensional embedding of the data based on these affine approximations.

 We then classified the different clusters in the low-dimensional embedding of the data.

This work can also be extended towards other NLDR techniques namely: Laplacian Eigenmaps, Hessian LLE etc..
